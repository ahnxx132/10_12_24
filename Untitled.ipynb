```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
# # Data loading
# file_path = Path("10_11_24.csv")
# emulsion_df = pd.read_csv(file_path)
# emulsion_df

# Load the CSV and Excel files
aqueous_layer_df = pd.read_csv('Aqueous_layer.csv')
break_df = pd.read_excel('Break.xlsx')
coagulation_df = pd.read_excel('Coagulation.xlsx')
coalescence_df = pd.read_excel('Coalescence.xlsx')

# Display first few rows of each dataset for context
aqueous_layer_df.head()
```

```python
# Function to clean percentage columns
def clean_percentage_columns(df, percentage_columns):
    for col in percentage_columns:
        df[col] = df[col].str.replace('%', '').astype(float)
    return df

# Cleaning the aqueous layer percentages
aqueous_layer_df = clean_percentage_columns(aqueous_layer_df, ['LS', 'JMSP', 'ESP', 'CO5', 'TeginOV', '2 H', 'Day 1', 'Day 2'])

# Cleaned data
aqueous_layer_df.head()
```

```python
# Merging datasets on 'Unnamed: 0' (row index)
merged_df = pd.merge(aqueous_layer_df, break_df.drop(columns=['LS', 'JMSP', 'ESP', 'CO5', 'TeginOV']), left_on='Unnamed: 0', right_on='Unnamed: 0', suffixes=('_aqueous', '_break'))
merged_df = pd.merge(merged_df, coagulation_df.drop(columns=['LS', 'JMSP', 'ESP', 'CO5', 'TeginOV']), on='Unnamed: 0', suffixes=('', '_coag'))
merged_df = pd.merge(merged_df, coalescence_df.drop(columns=['LS', 'JMSP', 'ESP', 'CO5', 'TeginOV']), on='Unnamed: 0', suffixes=('', '_coal'))

# Standardize column names
merged_df.columns = merged_df.columns.str.replace(' ', '_').str.replace('_break', '_Break').str.replace('_coag', '_Coag').str.replace('_coal', '_Coal')

# Display merged data
merged_df.head()
```

```python
merged_df.to_csv('merged_df1.csv')
```

```python
# Step 1: Ranking based on aqueous layer percentage (lower is better)
merged_df['avg_aqueous_layer'] = merged_df[['2_H', 'Day_1_aqueous', 'Day_2_aqueous']].mean(axis=1, skipna=True)
aqueous_layer_rank = merged_df['avg_aqueous_layer'].rank(method='min')

# Step 2: Counting occurrences of "N" (good performance)
merged_df['N_count'] = merged_df[['2H', 'Day_1_Break', 'Day_2_Break', '2H_Coag', 'Day_1', 'Day_2', '2H_Coal', 'Day_1_Coal', 'Day_2_Coal']].apply(lambda row: row.str.count('N').sum(), axis=1)

# Step 3: Combining the rankings (lower aqueous layer rank and higher N_count rank)
merged_df['overall_rank'] = aqueous_layer_rank + (merged_df['N_count'].max() - merged_df['N_count'])

# Sorting the dataframe by overall rank
ranked_df = merged_df.sort_values(by='overall_rank').reset_index(drop=True)

# Display ranked data
ranked_df.head()
```

```python
# Plotting the stability trends over time for top-ranked compositions
top_compositions = ranked_df.head(5)

# Melting the dataframe for time-based visualization
stability_trends = top_compositions.melt(id_vars=['Unnamed:_0', 'N_count', 'avg_aqueous_layer'], 
                                         value_vars=['2_H', 'Day_1_aqueous', 'Day_2_aqueous'],
                                         var_name='Time', value_name='Aqueous_Layer_Percentage')

# Plotting the stability trends over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=stability_trends, x='Time', y='Aqueous_Layer_Percentage', hue='Unnamed:_0', marker='o')
plt.title('Aqueous Layer Stability Trends Over Time (Top 5 Compositions)')
plt.ylabel('Aqueous Layer Percentage (%)')
plt.xlabel('Time')
plt.legend(title='Composition ID')
plt.show()
```

```python
# Plotting the overall performance distribution
plt.figure(figsize=(10, 6))
sns.histplot(ranked_df['overall_rank'], bins=10, kde=True)
plt.title('Overall Performance Distribution')
plt.xlabel('Overall Rank (Lower is Better)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation analysis on relevant columns
correlation_data = ranked_df[['LS', 'JMSP', 'ESP', 'CO5', 'TeginOV', 'avg_aqueous_layer', 'N_count', 'overall_rank']]

# Correlation matrix
correlation_matrix = correlation_data.corr()

# Plotting the correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title('Correlation Matrix for Emulsifier Components and Performance Metrics')
plt.show()
```

```python
# Creating a heatmap using % aqueous layer and N counts as stability metrics across all rows

# Selecting relevant columns for the heatmap
heatmap_data = ranked_df[['Unnamed:_0', 'avg_aqueous_layer', 'N_count']].set_index('Unnamed:_0')

# Plotting the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title('Heatmap of Emulsion Stability: Aqueous Layer % and N Counts')
plt.ylabel('Emulsion Sample (Row)')
plt.xlabel('Stability Metrics')
plt.show()
```

```python
# # Generating a table that shows ID, N counts, % aqueous layer, and overall stability rank
# stability_table = ranked_df[['Unnamed:_0', 'N_count', 'avg_aqueous_layer', 'overall_rank']].copy()
# stability_table.columns = ['ID', 'N Count', 'Average Aqueous Layer (%)', 'Overall Stability Rank']


# # Show the first few rows of the table
# stability_table


# Correcting the overall stability rank for all rows using the method of summing the aqueous layer rank and N count rank

# Recalculate the ranks for aqueous layer and N count
ranked_df['Aqueous_Layer_Rank'] = ranked_df['avg_aqueous_layer'].rank(method='min')
ranked_df['N_Count_Rank'] = ranked_df['N_count'].rank(ascending=False, method='min')

# Calculate the new overall rank by summing the ranks
ranked_df['overall_rank'] = ranked_df['Aqueous_Layer_Rank'] + ranked_df['N_Count_Rank']

# Regenerate the Emulsion Stability Table with the corrected ranks
stability_table_corrected = ranked_df[['Unnamed:_0', 'N_count', 'avg_aqueous_layer', 'overall_rank']].copy()
stability_table_corrected.columns = ['ID', 'N Count', 'Average Aqueous Layer (%)', 'Overall Stability Rank']


# Show the first few rows of the corrected table
stability_table_corrected.head(14)
```

```python
# Generating a new Emulsion Stability Table including only the emulsifiers used for each row

# Function to filter and return only the emulsifiers used (non-zero values)
def get_emulsifiers_used(row):
    emulsifiers = []
    if row['LS'] > 0:
        emulsifiers.append(f"LS: {row['LS']}%")
    if row['JMSP'] > 0:
        emulsifiers.append(f"JMSP: {row['JMSP']}%")
    if row['ESP'] > 0:
        emulsifiers.append(f"ESP: {row['ESP']}%")
    if row['CO5'] > 0:
        emulsifiers.append(f"CO5: {row['CO5']}%")
    if row['TeginOV'] > 0:
        emulsifiers.append(f"TeginOV: {row['TeginOV']}%")
    return ', '.join(emulsifiers)

# Apply the function to create a new column listing only the emulsifiers used in each row
ranked_df['Emulsifiers_Used'] = ranked_df.apply(get_emulsifiers_used, axis=1)

# Creating the new Emulsion Stability Table with the corrected ranks and emulsifiers used
stability_table_with_emulsifiers = ranked_df[['Unnamed:_0', 'N_count', 'avg_aqueous_layer', 'overall_rank', 'Emulsifiers_Used']].copy()
stability_table_with_emulsifiers.columns = ['ID', 'N Count', 'Average Aqueous Layer (%)', 'Overall Stability Rank', 'Emulsifiers Used']


# Show the first few rows of the updated table
stability_table_with_emulsifiers.head(14)
```
